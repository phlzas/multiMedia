```python
import cv2
import numpy as np
import imutils as imu
import matplotlib.pyplot as plt
from IPython.display  import Audio
from skimage.feature import local_binary_pattern
```

```python
image  =  cv2.imread("car.jpeg", cv2.IMREAD_GRAYSCALE)

raduis = 1
n_points = 8 * raduis
lbp = cv2.localBinaryPattern(image, n_points, raduis, method="uniform")
lbp_histogram = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
lbp_hist = lbp_histogram.astype("float")
lbp_hist /= (lbp_hist.sum() + 1e-6)



plt.figure(figsize=(30,20))
plt.subplot(1,7,1)
plt.imshow(image, cmap="gray") 
plt.title("original image")
plt.axis("off")     
plt.subplot(1,7,6)
plt.bar(np.arange(0, len(lbp_hist)), lbp_hist, width=0.5)
plt.title("LBP Histogram")
plt.xlabel("LBP value")
plt.ylabel("Frequency")
plt.show()
```
