```python
import cv2
import numpy as np
import imutils as imu
import matplotlib.pyplot as plt
from IPython.display  import Audio
```

```python
image  =  cv2.imread("car.jpeg", cv2.IMREAD_GRAYSCALE)
gussian_temp = np.zeros_like(image, dtype= np.uint8)

gussian_noise = cv2.randn(gussian_temp ,128,20)
uniform_noise  = cv2.randu(gussian_temp,0,255)

gussian_flitered = cv2.add(gussian_noise,image)
uniform_filterd  = cv2.add(uniform_noise,image )


plt.figure(figsize=(22,12))
plt.subplot(1,5,1)
plt.imshow(image, cmap="gray") 
plt.title("orginal")
plt.axis("off")
plt.subplot(1,5,2)
plt.imshow(gussian_noise,cmap="gray") 
plt.title("gussian noise")
plt.axis("off")
plt.subplot(1,5,3)
plt.imshow(gussian_flitered, cmap="gray") 
plt.title("gussan noise applied")
plt.axis("off")
plt.subplot(1,5,4)
plt.imshow(uniform_noise,cmap="gray") 
plt.title("uniform noise")
plt.axis("off")
plt.subplot(1,5,5)
plt.imshow(uniform_filterd, cmap="gray") 
plt.title("uniform noise applied")
plt.axis("off")
```
