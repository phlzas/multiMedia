```python
import cv2
import numpy as np
import imutils as imu
import matplotlib.pyplot as plt
from IPython.display  import Audio
```

```python
image  =  cv2.imread("car.jpeg", cv2.IMREAD_GRAYSCALE)
gussian_temp_black = np.zeros_like(image, dtype= np.uint8)
gussian_temp_white = np.ones_like(image, dtype= np.uint8)


gussian_noise_black= cv2.randn(gussian_temp_black ,128,10)
gussian_noise_white = cv2.randn( gussian_temp_white,128,10)
uniform_noise  = cv2.randu(gussian_temp_black,0,255)

gussian_flitered_black = cv2.add(gussian_noise_black,image)
gussian_flitered_white = cv2.add(gussian_noise_white,image)
uniform_filterd  = cv2.add(uniform_noise,image )

cv2.imwrite("gussian_noise_black.jpg", gussian_flitered_white)
plt.figure(figsize=(30,20))
plt.subplot(1,7,1)
plt.imshow(image, cmap="gray") 
plt.title("orginal")
plt.axis("off")
plt.subplot(1,7,2)
plt.imshow(gussian_noise_black,cmap="gray") 
plt.title("gussian noise")
plt.axis("off")
plt.subplot(1,7,3)
plt.imshow(gussian_flitered_black, cmap="gray") 
plt.title("gussan noise applied")
plt.axis("off")
plt.subplot(2,7,2)
plt.imshow(gussian_noise_white, cmap="gray") 
plt.title("gussian noise white")
plt.axis("off")
plt.subplot(2,7,3)
plt.imshow(gussian_flitered_white, cmap="gray") 
plt.title("gussan noise applied white")
plt.axis("off")
plt.subplot(1,7,4)
plt.imshow(uniform_noise,cmap="gray") 
plt.title("uniform noise")
plt.axis("off")
plt.subplot(1,7,5)
plt.imshow(uniform_filterd, cmap="gray") 
plt.title("uniform noise applied")
plt.axis("off")
```
